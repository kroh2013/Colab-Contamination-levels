```python
# download data
csv_path = os.path.join("/content/drive/MyDrive/S_Mn_50ppb.csv")
csv_path
```

```python
# import file as pandas df

mn_50ppbb = pd.read_csv(csv_path,
                              parse_dates=["Date"],
                              index_col=["Date"],
                              na_values=["999.99"])

mn_50ppbb
```

```python
#Plot dataframe
mn_50ppbb.plot(y="Result",
               title= "Surface Levels in the San Juan River")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,5))

ax.plot(mn_50ppbb.index.values,
           mn_50ppbb["Result"],
           color= "blue")

# Set title and labels for axes
ax.set(ylabel="Manganese Levels (in ug/L)",
       facecolor= "lightgray",
       title="Manganese Levels in surface water\n of the San Juan River")

# Plot axhline to mark good values
plt.axhline(5000, color = "red", ls="dotted")
ax.annotate("High Level", xy=(0.6,0.4), xycoords="figure fraction",
            xytext=(0.70,0.67), textcoords="figure fraction",
            va="center", ha= "left",
            arrowprops=dict(arrowstyle="simple", facecolor="red"))

#Rotate tick marks on x-axis
plt.setp(ax.get_xticklabels(), rotation=45)

fig.savefig('Mnplot.png')
#plt.show()
```
